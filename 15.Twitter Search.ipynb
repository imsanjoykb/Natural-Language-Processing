## 0. Helper method:

```python
def strip_whitespace(s):
    return re.sub(r'\s\s+', ' ', s)
```

## 1. Searching by keywords:

```python
from TwitterSearch import *
try:
    tso = TwitterSearchOrder()
    tso.set_keywords(['car', 'drive'])
    tso.set_language('en')
    tso.set_include_entities(False) # We don't want the entity information to be returned.
    # For more options: https://twittersearch.readthedocs.io/en/latest/advanced_usage_tso.html (The most useful in my opinion are set_locale() and set_result_type())
    # To exclude retweets and replies: https://github.com/ckoepp/TwitterSearch/issues/37
    # To retrieve non-truncated tweets (full tweets): https://github.com/ckoepp/TwitterSearch/issues/35
    
    # Creating TwitterSearch object
    ts = TwitterSearch(
        consumer_key = 'INSERT_YOUR_OWN_HERE',
        consumer_secret = 'INSERT_YOUR_OWN_HERE',
        access_token = 'INSERT_YOUR_OWN_HERE',
        access_token_secret = 'INSERT_YOUR_OWN_HERE'
     )

    for tweet in ts.search_tweets_iterable(tso):
        print( '@%s tweeted: %s' % (tweet['user']['screen_name'], strip_whitespace(tweet['text']))

except TwitterSearchException as e:
    print(e)
    
finally:
    if not save_file.closed:
        save_file.close()
```
